```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('properties_FE.csv').drop_duplicates()
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df1 = df[df['property_type']=='house']
df1['area'] = df1['area'].apply(lambda x:x * 10.7639)
df.update(df1)
```

```python
sns.distplot(df['area'])
```

```python
sns.boxplot(df['area'], orient = 'h')
```

```python
df[df['area']==53.819500]
```

```python
df[df['area']<300]['built_up_area']
```

```python
df[df['area']<300]['built_up_area'].shape
```

```python
df['area'] = np.where(np.abs(df['area'] - df['built_up_area'])>100,
                      np.maximum(df['built_up_area'],df['area']),df['area'])
```

```python
df['area'].describe()
```

```python
df[df['area']<12000].shape
df[df['area']<12000]['area'].describe()
```

```python
df = df[df['area']<12000]
```

## Check in Price Column

```python
sns.distplot(df['price'])
```

```python
sns.boxplot(x=df['price'])
```

```python
# outliers according to IQR
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_df = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
print(outliers_df.shape)
outliers_df['price'].describe()
```

```python
outliers_df.sort_values('price',ascending=False).head(10)
```

```python
outliers_df.to_csv('outliers.csv',index=False)
```
