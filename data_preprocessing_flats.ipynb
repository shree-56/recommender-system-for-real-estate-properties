```python
import numpy as np
import pandas as pd
import re
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('flats.csv')
df.shape
```

```python
df.info()  #every column is an object
```

```python
df.sample(6)
```

```python
# Checking for duplicates
df.duplicated().sum()
```

```python
# Dropping link and prop_id because they are of no use in this analysis
df.drop(columns=['link','property_id'], inplace = True) 
```

```python
df.price.value_counts()
```

```python
df[df['price'].isnull()]
```

```python
# prices are in lac, crore and price on request. Convert all into crores
#dropping the ones with Price on Request category

df = df[df['price']!= 'Price on Request']
def convert_to_crore(value):
    if type(value) == float:
        return value
    else:
        amt = str(value).split(" ")[0]
        unit = str(value).split(" ")[1]
        if unit != 'Crore':
            return round(float(amt)/100,2)
        else:
            return round(float(amt),2)

df['price'] = df['price'].apply(convert_to_crore)     
```

```python
df['price_per_sqft'].isnull().sum()
```

```python
df.dropna(subset=['price_per_sqft'],inplace = True)
```

```python
df['price_per_sqft'].value_counts()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].apply(lambda x: re.sub(r'/.+','',str(x).replace(',','').replace('₹',''))).astype('float')
df['price_per_sqft']
```

```python
# Calculate area from total price and price_per_sqft
df.insert(loc=4,column='area',value=round((df['price']*10000000)/df['price_per_sqft']))
```

```python
df.sample(6)
```

```python
# There are some ratings at the end of society name, thus cleaning is required
df['society'] = df['society'].fillna('independent')
df['society'] = df['society'].apply(lambda x: re.sub(r'\d+(\.\d)?\s?★?','',x).strip().lower())
# checking if cleaning done properly!
#df[df['society'].str.contains('★')]
```

```python
df.sample(6)
```

```python
df.info()
```

```python
df.dropna(subset=['bedRoom'],inplace=True)
df['bedRoom'].isnull().sum()
df['bathroom'].isnull().sum()
```

```python
df['bedRoom'] = df['bedRoom'].str.split(" ").str.get(0).astype('int')
df['bedRoom'].value_counts()
```

```python
df['bathroom'].value_counts()
```

```python
df['bathroom'] = df['bathroom'].str.split(" ").str.get(0).astype('int')
df['bathroom'].value_counts()
```

```python
df['balcony'].value_counts()
```

```python
# There is a category name 3+ balconies. Thus we can't convert this column to int.
df['balcony'] = df['balcony'].str.split(" ").str.get(0).str.replace('No','0')
df['balcony'].value_counts()
```

```python
df['additionalRoom'].isnull().sum()
```

```python
df['additionalRoom'] = df['additionalRoom'].fillna('not available').str.lower()
```

```python
df['floorNum'] = df['floorNum'].str.split(' ').str.get(0).replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract(r'(\d+)')
```

```python
df['facing'].fillna('not available',inplace=True)
```
